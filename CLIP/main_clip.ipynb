```python
import os
import torch
import numpy as np
from PIL import Image
from fashion_clip import fashion_clip
from scipy.spatial.distance import cosine, euclidean
import matplotlib.pyplot as plt
import pandas as pd
import ast

# Définir les chemins
DAM_FOLDER = "../DAM_modified"
TEST_FOLDER = "../test_image_modified"
MATCHING_FILE = "../matching.csv"
```

```python
# Initialiser FashionCLIP
clip_model = fashion_clip.FashionCLIP('fashion-clip')
print("FashionCLIP model loaded")
```

```python
def extract_embeddings(folder_path, clip_model, output_file, batch_size=32):
    """
    Extract embeddings using FashionCLIP
    
    Args:
        folder_path (str): Path to the folder containing images
        clip_model: FashionCLIP model instance
        output_file (str): Path where embeddings will be saved
        batch_size (int): Batch size for processing images (default: 32)
    """
    embeddings = {}
    image_files = [f for f in os.listdir(folder_path) if f.endswith('.png')]
    
    for file in image_files:
        try:
            image_path = os.path.join(folder_path, file)
            image = Image.open(image_path).convert('RGB')
            
            # Encode the image with FashionCLIP, now including batch_size
            embedding = clip_model.encode_images([image], batch_size=batch_size)
            if isinstance(embedding, torch.Tensor):
                embedding = embedding.detach().cpu().numpy()
            
            file_name = os.path.splitext(file)[0]
            embeddings[file_name] = embedding
            print(f"Processed {file}")
        except Exception as e:
            print(f"Error processing {file}: {e}")
    
    np.save(output_file, embeddings)
    print(f"Embeddings saved to {output_file}")
    return embeddings

# Extract embeddings with batch_size parameter
print("Extracting training embeddings...")
embeddings_train = extract_embeddings(DAM_FOLDER, clip_model, "embeddings_fashionclip_train.npy", batch_size=32)
print("\nExtracting test embeddings...")
embeddings_test = extract_embeddings(TEST_FOLDER, clip_model, "embeddings_fashionclip_test.npy", batch_size=32)
```

```python
def calculate_distance(emb1, emb2, metric='cosine'):
    """Calculate distance between two embeddings using specified metric"""
    # Aplatir les embeddings si nécessaire
    emb1_flat = emb1.flatten()
    emb2_flat = emb2.flatten()
    
    if metric == 'cosine':
        return cosine(emb1_flat, emb2_flat)
    elif metric == 'l1':
        return np.sum(np.abs(emb1_flat - emb2_flat))
    elif metric == 'l2':
        return euclidean(emb1_flat, emb2_flat)
    else:
        raise ValueError(f"Unsupported metric: {metric}")

def find_closest_matches(embeddings_test, embeddings_train, metric='cosine'):
    """Find closest matches using specified distance metric"""
    closest_embeddings = {}
    
    for test_key, test_embedding in embeddings_test.items():
        distances = []
        
        for train_key, train_embedding in embeddings_train.items():
            distance = calculate_distance(test_embedding, train_embedding, metric)
            distances.append([train_key, distance])
        
        distances.sort(key=lambda x: x[1])
        new_dist = [dist for dist in distances if not dist[0].startswith('flip_')]
        closest_embeddings[test_key] = new_dist[:5]
    
    return closest_embeddings
```

```python
def calculate_accuracy(closest_embeddings, matching_df):
    """Calculate Top-1, Top-3, and Top-5 accuracy"""
    true_count_1 = 0
    true_count_3 = 0
    true_count_5 = 0
    total = len(closest_embeddings)
    
    for test_key, closest_matches in closest_embeddings.items():
        match_liste = matching_df.loc[matching_df['test'] == test_key]['DAM'].iloc[0]
        
        # Top-1
        if closest_matches[0][0] in match_liste:
            true_count_1 += 1
            print(f"Top-1 Match for {test_key}")
        
        # Top-3
        for match in closest_matches[:3]:
            if match[0] in match_liste:
                true_count_3 += 1
                print(f"Top-3 Match for {test_key}")
                break
        
        # Top-5
        for match in closest_matches[:5]:
            if match[0] in match_liste:
                true_count_5 += 1
                print(f"Top-5 Match for {test_key}")
                break
    
    return {
        'top1': true_count_1 / total,
        'top3': true_count_3 / total,
        'top5': true_count_5 / total
    }
```

```python
def visualize_matches(test_key, closest_matches, test_folder, train_folder):
    """Visualize test image and its closest matches"""
    fig, axes = plt.subplots(1, 4, figsize=(15, 4))
    
    # Test image
    test_path = os.path.join(test_folder, f"{test_key}.png")
    test_img = Image.open(test_path)
    axes[0].imshow(test_img)
    axes[0].set_title("Test Image")
    axes[0].axis("off")
    
    # Top 3 matches
    for i, (match_key, distance) in enumerate(closest_matches[:3], 1):
        match_path = os.path.join(train_folder, f"{match_key}.png")
        match_img = Image.open(match_path)
        axes[i].imshow(match_img)
        axes[i].set_title(f"Match {i}\nDistance: {distance:.3f}")
        axes[i].axis("off")
    
    plt.tight_layout()
    plt.show()
```

```python
# Charger les données de matching
matching_df = pd.read_csv(MATCHING_FILE, sep=";")
matching_df['DAM'] = matching_df['DAM'].apply(ast.literal_eval)

# Calculer les similarités avec différentes métriques
metrics = ['cosine', 'l1', 'l2']
results = {}

for metric in metrics:
    print(f"\nCalculating similarities using {metric} distance...")
    closest_matches = find_closest_matches(embeddings_test, embeddings_train, metric)
    accuracy = calculate_accuracy(closest_matches, matching_df)
    results[metric] = accuracy
    
    print(f"\nResults for {metric} distance:")
    print(f"Top-1 Accuracy: {accuracy['top1']:.3f}")
    print(f"Top-3 Accuracy: {accuracy['top3']:.3f}")
    print(f"Top-5 Accuracy: {accuracy['top5']:.3f}")
    
    # Visualiser quelques exemples pour la distance cosinus
    if metric == 'cosine':
        print("\nVisualizing first 5 test images and their matches...")
        for test_key in list(closest_matches.keys())[:5]:
            visualize_matches(test_key, closest_matches[test_key], TEST_FOLDER, DAM_FOLDER)
```
