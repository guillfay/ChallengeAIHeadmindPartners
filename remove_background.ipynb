### Second code : Supprimer totalement le bg (fond png) et enregistrer les nouvelles images

```python
from rembg import remove
from PIL import Image
import os

# Dossier contenant les images originales
input_dir = r'C:\Users\pierr\OneDrive - Groupe ESIEA\Desktop\PyDump\EDC headmind\ChallengeAIHeadmindPartners\data\test_image_headmind'

# Dossier de sortie pour les images transformées
output_dir = r'C:\Users\pierr\OneDrive - Groupe ESIEA\Desktop\PyDump\EDC headmind\ChallengeAIHeadmindPartners\data\processed_data'

os.makedirs(output_dir, exist_ok=True)

# Fonction pour gérer la transparence
def remove_transparency(image, background_color=(255, 255, 255)):
    if image.mode in ('RGBA', 'LA') or (image.mode == 'P' and 'transparency' in image.info):
        alpha = image.convert('RGBA').split()[-1]
        background = Image.new("RGB", image.size, background_color)
        background.paste(image, mask=alpha)
        return background
    else:
        return image.convert("RGB")

# Lister les fichiers du dossier
image_files = [f for f in os.listdir(input_dir) if f.endswith(('jpg', 'jpeg', 'png'))]

for img_name in image_files:
    img_path = os.path.join(input_dir, img_name)
    
    # Supprimer le fond
    input_image = Image.open(img_path)
    output_image = remove(input_image)
    
    # Enregistrer en PNG pour conserver la transparence
    if output_image.mode == 'RGBA':
        output_path = os.path.join(output_dir, img_name.split('.')[0] + '.png')
        output_image.save(output_path, format='PNG')
    
    # Si JPEG est préféré, convertir en RGB avec fond blanc
    else:
        output_image = remove_transparency(output_image)
        output_path = os.path.join(output_dir, img_name)
        output_image.save(output_path, format='JPEG')

    print(f"Processed and saved: {output_path}")
```
